# Script to quantify variations minimum of the Ae. tauschii accesison by windows from IBSpy combined tables by chromsome 

```python
import pandas as pd
import numpy as np
```

## Functions

```python
def select_genotypes(df, sample_list):
    filter_df = df.set_index(['seqname', 'start', 'end']).filter(items=sample_list, axis=1)
    return filter_df.reset_index()

def file_to_list(file):
    file_l = pd.read_csv(file)['genotype'].tolist()
    return file_l

def get_min(df, list_1, list_2):
    monoc_df = select_genotypes(df, list_1)
    out_df = select_genotypes(df, list_2)
    out_df.insert(3, 'aet_min', monoc_df.iloc[:, 3:].min(axis=1))
    return out_df

def filter_min(df, filter_v):
    for value in filter_v:
        c_pos = filter_v.index(value)
        df.insert(4+c_pos, f'aet_min_{value}', np.where(min_df['aet_min']<= value, 'TRUE','FALSE'))
    return df
```

##  Arguments

```python
window = 50000
reference = 'sy_mattis'
data_path = '/Users/quirozc/Documents/PhD_2020/16_watseq/0_IBSpy_data/'
```

## Input IBSpy variaitons table data

```python
in_file = pd.read_csv(f'{data_path}/{reference}_combined_queries_{window}w.tsv.gz', delimiter='\t')
in_file
```

## Filter by sub-genome A

```python
by_genome = in_file[in_file['seqname'].str.contains('D')].copy()
by_genome
```

## Get Ae. tauschii minimum variaitons count by window

```python
aet_list = file_to_list('Ae_tauschii_genotypes.tsv')
genotypes_list = file_to_list('genotypes.tsv')
min_df = get_min(by_genome, aet_list, genotypes_list)
min_df
```

## Assign <30 or 120 regions

```python
values_f = [30, 120]
min_value_df = filter_min(min_df, values_f)
min_value_df
```

## Split by chromosome & save tables

```python
chromosomes = ('chr1D', 'chr2D', 'chr3D', 'chr4D', 'chr5D','chr6D','chr7D')

for chromosome in chromosomes:
    chrom_ind = min_value_df[min_value_df['seqname'].str.contains(chromosome)].copy()
    chrom_ind.to_csv(f'{reference}_{chromosome}_variations_all_genotypes.tsv.gz', sep='\t', index=False)
```
