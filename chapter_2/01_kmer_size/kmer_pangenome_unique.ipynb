```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
from matplotlib.ticker import ScalarFormatter
```

```python
in_db = pd.read_csv('compiled_kmer_size_histo.tsv', delimiter='\t')
in_db
```

```python
in_db_filter = in_db[in_db['frequency'] == 1]
in_db_filter
```

```python
# in_db_filter.set_index('kmer_size')
```

```python
def melt_df(df):
    df_m = pd.melt(df, id_vars=['kmer_size','frequency'],
                   value_vars=df.columns[2:],
                   var_name='query',
                   value_name='count')
    return df_m

test = melt_df(in_db_filter)
test
```

```python
kmer_sizes = list(test['kmer_size'].unique())
kmer_sizes
```

```python
font_s = 10
font_w = "bold"

sns.set_theme(style="whitegrid")
f, ax = plt.subplots(figsize=(10, 3.4))

sns.lineplot(data=test, x="kmer_size", y="count", hue="query", markers=True, lw=2)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
# f.axes[0].yaxis.set_major_formatter(ScalarFormatter())
# plt.yscale('symlog', base=2)

# plt.title(f'Pangenome unique k-mers frequency', fontweight=font_w, fontsize= font_s)
plt.ylabel('Count (Log)', fontsize = font_s, fontweight=font_w)

plt.xticks([10,13,15,17,19,21,25,31,35,41,51,61,71,81,91,101], fontsize = font_s)
ax.set_xticklabels(ax.get_xticklabels(), fontsize = font_s)
plt.yticks(fontsize = font_s)
plt.xlabel('k-mer size', fontsize = font_s, fontweight=font_w)
# sns.despine(offset=1, trim=False)
```
