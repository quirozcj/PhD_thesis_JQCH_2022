```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
```

# Use the combined file created in the HPC

```python
# Vilin plot
path = '/Volumes/quirozj/09_watseq/14_kmer_histo/histo/'
in_file = pd.read_csv(path+'kmer_histo_combined.tsv', delimiter='\t')
in_file
```

```python
su_df = in_file.iloc[:40, :].set_index('frequency').T
su_df
```

```python
# https://seaborn.pydata.org/examples/wide_form_violinplot.html
def vilon_plot(df, figsize):
    font_w = 'bold'
    font_s = 20
    sns.set_theme(style="whitegrid")
    f, ax = plt.subplots(figsize=(15, 4))
#     f, ax = plt.subplots(figsize=figsize)
    sns.set_theme(style="whitegrid")
    sns.violinplot(data=df, palette="Set3", bw=.2, cut=1, linewidth=1)
    # ax.set(ylim=(-.7, 1.05))
#     plt.xscale('symlog', base=2)
    plt.xticks(list(range(0, 50, 2)),fontsize = 20)
    plt.yticks(fontsize = 20)
    plt.title(f'Wheat WatSeq k-mer distribution', fontweight=font_w, fontsize= font_s)
    plt.ylabel('Number of distinct k-mers', fontsize = font_s, fontweight=font_w)
    plt.xlabel('Counts per k-mer (depth)', fontsize = font_s, fontweight=font_w)
#     ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
#     sns.despine(left=True, bottom=True)
```

```python
# https://seaborn.pydata.org/examples/wide_form_violinplot.html
def box_plot(df, figsize):
    font_w = 'bold'
    font_s = 15
    sns.set_theme(style="whitegrid")
    f, ax = plt.subplots(figsize=figsize)
#     f, ax = plt.subplots(figsize=figsize)
    sns.boxplot(data=df, palette="Set3", linewidth=1)
    # ax.set(ylim=(-.7, 1.05))
#     plt.xscale('symlog', base=2)
    plt.xticks(list(range(0, 40, 1)),fontsize = 10)
    plt.yticks(fontsize = 10)
    plt.title(f'Wheat WatSeq k-mer distribution', fontweight=font_w, fontsize= font_s)
    plt.ylabel('Number of distinct k-mers', fontsize = font_s, fontweight=font_w)
    plt.xlabel('Counts per k-mer (depth)', fontsize = font_s, fontweight=font_w)
#     ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
#     sns.despine(left=True, bottom=True)
```

```python
# plot a the firts 20 freauency
figsize = (10, 3)
box_plot(su_df, figsize)
```

```python
# plot a the firts 20 freauency
figsize = (21, 6)
vilon_plot(su_df, figsize)
```

```python
# plot all data
all_data = in_file.iloc[:50, :].set_index('frequency').T
vilon_plot(all_data, figsize)
```
